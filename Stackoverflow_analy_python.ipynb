```python
import pandas
import numpy as np
import matplotlib as mpl           # 그래프를 그리는 패키지
import matplotlib.pyplot as plt    # 그래프를 그리는 패키지
from pandas import read_csv
```

```python
stackoverflow = read_csv('/Users/choejiho/modified_survey_results.csv')
stackoverflow_dsc_converted_comp = stackoverflow.sort_values('ConvertedCompYearly', ascending=False)
stackoverflow_dsc_converted_comp
```

```python
converted_comp = stackoverflow_dsc_converted_comp['ConvertedCompYearly']
# converted_comp = [110.0, 20.0, 30.0, 40.0, 50.0, 610.0, 70.0, 80.0, 90.0, 100.0, 110.0]
```

```python

# 구간별 빈도수 계산
bins = np.arange(100, 5000000, 100000)  # 0부터 100까지 10의 단위로 구간을 나눔
bin_values, _ = np.histogram(converted_comp, bins=bins)

# 그래프 그리기
plt.bar(range(len(bin_values)), bin_values)

# x축 레이블과 제목 추가
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.title('Distribution of ConvertedCompYearly')

# x축 레이블 설정
x_labels = [f'{bins[i]}-{bins[i+1]}' for i in range(len(bins)-1)]
plt.xticks(range(len(bin_values)), x_labels, rotation=45)

# 그래프 보여주기
plt.show()
```
